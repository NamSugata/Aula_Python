```python
import math as mt
import numpy as np

beta = np.array([0.1, 0.2]) # Gerando um beta qualquer

def gamma(beta):   
    """Função que toma o parametro beta como argumento e retorna gamma"""
    nova_lista = []
    for (i, valor) in enumerate(beta):
        novo_elemento = 1 / (mt.sqrt(1 - valor**2))
        nova_lista.append(novo_elemento)
    return nova_lista

print(gamma(beta))
```

```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt

# vetor de beta
beta = np.arange(0,0.995,0.01)

# gamma calculado a partir de Beta
def gamma(beta):   
    """Função que toma beta como argumento e retorna gamma"""
    nova_lista = []
    for (i, valor) in enumerate(beta):
        novo_elemento = 1/(mt.sqrt(1- valor**2))
        nova_lista.append(novo_elemento)
    return nova_lista

x = beta

fig, ax = plt.subplots(1,2,figsize=(12, 6), tight_layout=True )
ax[0].set_title('Gráfico γ vs. β')
ax[0].set_xlabel('β')
ax[0].set_ylabel('Fator de Lorentz (γ)')
ax[0].plot(x , gamma(beta), 'r-')

ax[1].set_title('Gráfico γβ vs β ')
ax[1].set_xlabel('β')
ax[1].set_ylabel('γβ')
ax[1].plot(x , beta*gamma(beta),'b-')

# comparar os plots
fig1, ax1 = plt.subplots(figsize=(12, 6), tight_layout=True )
ax1.plot(x, gamma(beta), 'r--', x, gamma(beta)*beta, 'b-')
ax1.legend(['γ', 'γβ'])
ax1.set_title('Gráfico comparação γβ e γ')
ax1.set_xlabel('β')
ax1.set_ylabel('Unidade Arbitrária')
ax1.grid()
ax1.fill_between(x,gamma(beta),gamma(beta)*beta, alpha= 0.2)
```
